# Merge

```python
df1 = pd.read_excel('./df1.xlsx')
df1
```

```python
df3 = pd.read_excel('./df3.xlsx')
df3
```

```python
# Merge On the index basis
df1.merge(df3,left_index=True,right_index=True)
```

```python
df2 = pd.read_excel('./df2.xlsx')
df2
```

```python
df1.merge(df2,on='NIC')
```

```python
df1.merge(df2,on='Nic',how='outer')
```

```python
df1.merge(df2,on='Nic',how='right')
```
